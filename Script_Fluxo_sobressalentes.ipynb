```python
#Lista de Clientes
#U-220 (DCCF)
Clients = df['Unidade'].unique()
Clients
```

```python
x = Clients[1]
clientedb = df[df['Unidade']== x]
#clientdb = df['Unidade']==x
clientedb
```

```python
pt_pn=clientedb.pivot_table(
    ['Qtde'],
    ['PN'],
    aggfunc="sum",
)
pt_pn.sort_values(by=['Qtde'],ascending=False)
```

```python
def recomendation(valor):
    if valor > 15:
        return 0.4
    elif valor > 1:
        return 0.5
    else:
        return 1

pt_pn['perc_Qtde_rec'] = pt_pn['Qtde'].map(recomendation)

pt_pn.sort_values(by=['Qtde'],ascending=False)
```

```python
#Top 10 itens
pt_pn["Qtde recomended"] = pt_pn['perc_Qtde_rec']*pt_pn['Qtde']
#pt_pn=pt_pn['Qtde recomended'].round(0)
pt_pn_top10 = pt_pn.head(10)
pt_pn_top10 = pt_pn_top10.sort_values(by=['Qtde'],ascending=False)
```

```python
# Recomentation for x%
Total=pt_pn['Qtde'].sum()
Total_rec=pt_pn_top10['Qtde recomended'].sum()

perc_supply = 100*Total_rec/Total
print(f"Ao comprar todos itens relacionados acima, é atigingido {perc_supply:.2f}% dos sobresalentes do seu parque")
```
